```python
from google.colab import drive
import pandas as pd
from pathlib import Path
from IPython.display import display
from PIL import Image
drive.mount('/content/drive')

# Base folders
base_folder = Path("/content/drive/MyDrive/LLMs/MLLM_CW/0730_GPTo3_Evaluation/evaluation_results/o3/")
```

```python
import json
import pandas as pd

# Load your JSON file
with open(f'{base_folder}/PortOperationalGuidance_eval_0731.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Convert to flattened format
flattened_data = []
for entry in data:
    row = {'id': entry['id']}
    for llm_key, result in entry['results'].items():
        suffix = llm_key.replace('Answer_', '')
        row[f'{suffix}_score'] = result.get('score')
        row[f'{suffix}_hallucinated'] = result.get('hallucinated')
        row[f'{suffix}_ReactionLevel'] = result.get('ReactionLevel')
    flattened_data.append(row)
```

```python
# Convert to DataFrame and save as CSV
df_PortOperationalGuidance_eval = pd.DataFrame(flattened_data)
df_PortOperationalGuidance_eval.to_csv(f'{base_folder}/PortOperationalGuidance_eval.csv', index=False)
```

## Port_level hazard

```python
# File suffixes
parts = ['part1', 'part2', 'part3', 'part4', 'part5']
base_filename = f'{base_folder}/Port-LevelHazardandForecastContext_eval_'  # e.g., input_part1.json

# Accumulate all flattened rows
flattened_data = []

for part in parts:
    filename = f'{base_filename}{part}.json'
    with open(filename, 'r', encoding='utf-8') as f:
        data = json.load(f)

    for entry in data:
        row = {'id': entry['id']}
        for llm_key, result in entry['results'].items():
            suffix = llm_key.replace('Answer_', '')
            row[f'{suffix}_score'] = result.get('score')
            row[f'{suffix}_hallucinated'] = result.get('hallucinated')
            row[f'{suffix}_ReactionLevel'] = result.get('ReactionLevel')
        flattened_data.append(row)

# Convert to DataFrame and export
df_Port_LevelHazardandForecastContext = pd.DataFrame(flattened_data)
df_Port_LevelHazardandForecastContext.to_csv(f'{base_folder}/Port-LevelHazardandForecastContext.csv', index=False)
```

## CycloneStatusAndForecastArea

```python
# File suffixes
parts = ['part1', 'part2', 'part3', 'part4', 'part5', 'part6', 'part7']
base_filename = f'{base_folder}/CycloneStatusAndForecastArea_eval_'  # e.g., input_part1.json

# Accumulate all flattened rows
flattened_data = []

for part in parts:
    filename = f'{base_filename}{part}.json'
    with open(filename, 'r', encoding='utf-8') as f:
        data = json.load(f)

    for entry in data:
        row = {'id': entry['id']}
        for llm_key, result in entry['results'].items():
            suffix = llm_key.replace('Answer_', '')
            row[f'{suffix}_score'] = result.get('score')
            row[f'{suffix}_hallucinated'] = result.get('hallucinated')
            row[f'{suffix}_ReactionLevel'] = result.get('ReactionLevel')
        flattened_data.append(row)

# Convert to DataFrame and export
df_Port_CycloneStatusAndForecastArea = pd.DataFrame(flattened_data)
df_Port_CycloneStatusAndForecastArea.to_csv(f'{base_folder}/CycloneStatusAndForecastArea.csv', index=False)
```
